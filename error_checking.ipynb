```python
from approximating_maxima import ApproximateMaxima
```

```python
tester=ApproximateMaxima([-5,6],4,sample_function='4*math.cos(x)')

tester.add_n_points(2,function_type='sample')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

fig, axs = plt.subplots(1,2,figsize=(12,6))
axs=axs.flatten()
xs=np.linspace(-5,6,100)
ys=4*np.cos(xs)
tester.graphs[-1].draw_to_axis(axs[0])
axs[0].plot(xs,ys, color='grey',alpha=.25)
axs[1].plot(xs,ys, color='grey',alpha=.25)

plt.show()
```
